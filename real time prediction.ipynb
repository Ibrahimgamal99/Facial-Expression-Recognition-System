```python
from mtcnn.mtcnn import MTCNN
from tensorflow import keras
from tensorflow.keras.preprocessing.image import load_img ,img_to_array
import cv2 as cv
import numpy as np
%matplotlib inline
```

```python
cap = cv.VideoCapture(0)
```

```python
label_map={0: 'angry',
 1: 'disgusted',
 2: 'fearful',
 3: 'happy',
 4: 'neutral',
 5: 'sad',
 6: 'surprised'}
```

```python
model = keras.models.load_model('wights.h5' , compile=False)
```

```python
def color(col):
    if col=='sad':
        return (42,59,144)
    elif col=='fearful':
        return (101, 155, 246)
    elif col=='happy':
        return (0, 255, 255)
    elif col=='angry':
        return (106, 39, 225)
    elif col=='disgusted':
        return (154,205,50)
    elif col=='neutral':
        return (238, 238, 238)
    elif col=='surprised':
        return (34, 231, 34)
```

```python
while True:
    # Read the frame
    ret, frame = cap.read()
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    detector = MTCNN()
    
    # detect faces in the image
    faces = detector.detect_faces(frame)
    
    for face in faces:
        x, y, w, h = face['box']
        crop=gray[y:y+h,x:x+w]
       
        if(type(faces)!=tuple):
            img = cv.resize(crop,(48,48))
            image=img_to_array(img).reshape(-1,48,48,1)
            image = image/255 
            predict=model.predict(image)
            predict=np.argmax(predict,axis = 1)
            cv.rectangle(frame, (x, y), (x+w, y+h),color(label_map[predict[0]]), 2)
            cv.putText(frame,label_map[predict[0]],(x-10, y-10),cv.FONT_HERSHEY_PLAIN,2, color(label_map[predict[0]]),2) 
   

    # Display
    cv.imshow('img', frame)
    # Stop if escape key is pressed
    if cv.waitKey(1) & 0xFF == ord('q'):
        break
# Release the VideoCapture object
cap.release()
cv.destroyAllWindows()
```
